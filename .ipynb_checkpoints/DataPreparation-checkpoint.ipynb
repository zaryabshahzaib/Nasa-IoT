## Data collection

```python
import numpy as np
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
train = pd.read_csv('data/train_FD001.txt', sep=" ", header=None)
test = pd.read_csv('data/test_FD001.txt', sep=" ", header=None)
test_labels = pd.read_csv('data/RUL_FD001.txt', sep=" ", header=None)
```

## Data Wrangling

#### As specified in the README file of the dataset, we make the column headers.

```python
unit_and_cycle = ["Unit_Number", "Cycle#"]
oper_setting = ["Operational_Setting_"+str(i) for i in range(1,4)]
sensors = ["Sensor_"+str(j) for j in range(1,22)] 
col_names = unit_and_cycle + oper_setting + sensors
print(col_names)
```

#### Checking for Missing Values in the Data

```python
train.isnull().sum(axis=0)
```

```python
test.isnull().sum(axis=0)
```

```python
test_labels.isnull().sum(axis=0)
```

#### Comment
As observed above, columns 26,27 have all NaN values in the train and test dataframes, and column 1 has all NaN in the test_labels dataframe. These columns were added by the pd.read_csv, they were not actually part of the dataset. So we will now remove these columns

```python
train.drop([26, 27], axis=1, inplace=True)
test.drop([26, 27], axis=1, inplace=True)
test_labels.drop([1], axis=1, inplace=True)
```

#### Now we add the column names to the train and testing dataframes

```python
train.columns = col_names
test.columns = col_names
```

#### The RUL (Target) column is missing from the train and test dataframes, so we have to calculate that for every entry and add it manually. 

```python
target_RUL=[]
for index, row in train.iterrows():
    temp_df = train.iloc[index+1:,:]
    if (not (temp_df[temp_df["Unit_Number"] ==row["Unit_Number"]]["Cycle#"]).empty):
        RUL = max(temp_df[temp_df["Unit_Number"] ==row["Unit_Number"]]["Cycle#"])
        target_RUL.append(int(RUL - row["Cycle#"]))
    else:
        target_RUL.append(0)
        
train['RUL'] = target_RUL
```

```python
target_RUL=[]
for index, row in test.iterrows():
    temp_df = test.iloc[index+1:,:]
    if (not (temp_df[temp_df["Unit_Number"] ==row["Unit_Number"]]["Cycle#"]).empty):
        RUL = max(temp_df[temp_df["Unit_Number"] ==row["Unit_Number"]]["Cycle#"])
        target_RUL.append(int(RUL - row["Cycle#"]) + int(test_labels.iloc[int(row["Unit_Number"] - 1)]))
    else:
        target_RUL.append(int(test_labels.iloc[int(row["Unit_Number"] - 1)]))
```

```python
test['RUL'] = target_RUL
```

### Data Scaling and Normalization

```python
from sklearn.preprocessing import StandardScaler

train_data_to_be_scaled = train.loc[:, oper_setting + sensors]
test_data_to_be_scaled = test.loc[:, oper_setting + sensors]
scaler = StandardScaler().fit(train_data_to_be_scaled)
scaler
```

```python
train_scaled = pd.DataFrame(scaler.transform(train_data_to_be_scaled))
test_scaled = pd.DataFrame(scaler.transform(test_data_to_be_scaled))
```

#### Now we add back the removed columns that we didnt want to be scaled.

```python
#Train
train_scaled.insert (0, 'Unit_Number', train["Unit_Number"])
train_scaled.insert (1, 'Cycle#', train["Cycle#"])
train_scaled.columns = col_names
train_scaled["RUL"] = train['RUL']
#Test
test_scaled.insert (0, 'Unit_Number', test["Unit_Number"])
test_scaled.insert (1, 'Cycle#', test["Cycle#"])
test_scaled.columns = col_names
test_scaled["RUL"] = test['RUL']
```

### Feature Engineering

As observed from the correlation matrix, and other plots above, we have concluded that none of the operational features, sensors[1, 5, 10, 16, 18, 19], and Unit_Number are correlated to the target (RUL). Hence we would be removing these features from our training and testing data, to finalize the data which would then be used as input to regression algorithims. 

```python
#Train
train_features = train_scaled.drop(["Unit_Number", "RUL"] + oper_setting + ["Sensor_"+str(i) for i in [1, 5, 10, 16, 18, 19]], axis=1)
train_labels = train_scaled["RUL"]
#Test
test_features = test_scaled.drop(["Unit_Number", "RUL"] + oper_setting + ["Sensor_"+str(i) for i in [1, 5, 10, 16, 18, 19]], axis=1)
test_labels = test_scaled["RUL"]
```

```python
print("Training Features shape is {}".format(train_features.shape))
print("Training Labels shape is {}\n".format(train_labels.shape))
print("Testing Features shape is {}".format(test_features.shape))
print("Testing Labels shape is {}".format(test_labels.shape))
```

### Recursive Feature Engineering

In addition to the features we already removed, there might be other features which are correlated to the target but do not contribute towards predicting the target. For recursive feature engineering, we would be starting with the features we have and trying out recursivley smaller sets of features and recording their resulting r2. I would be using the DecisionTreeRegressor as a reference model, this is because it provides a measure of feature_importance for the passed in features. After finding out important features with DecisionTreeRegressor, we will also confirm our hypothesis with KNearestRegressor.

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_squared_error
from sklearn import model_selection
import warnings
warnings.filterwarnings('ignore')
```

```python
tree = DecisionTreeRegressor().fit(train_features, train_labels)
```

### Visualizing Feature Importances

```python
tree = DecisionTreeRegressor().fit(train_features, train_labels)
plt.figure(figsize=(16, 9))
plt.scatter(train_features.columns, tree.feature_importances_)
plt.show()
```

#### Recursive Elimination

Starting from taking 1 feature, to taking all features, we will observe how r2 changes

```python
from sklearn.metrics import r2_score
features_dict = {}
num_features = []
features = []
mse_score_RT = []
r2_scores_RT = []

sorted_imp = np.sort(tree.feature_importances_)
reverse_sorted = sorted_imp[::-1]
for i in range(1, 17):
    features_tmp =[]
    for elem in (reverse_sorted[:i]):
        features_tmp.append(list(train_features.columns[np.where(tree.feature_importances_ == elem)]))
    flattened = [val for sublist in features_tmp for val in sublist]    
    tree_prime = DecisionTreeRegressor().fit(train_features.loc[:,flattened], train_labels)
    features_dict[i]  = flattened
    features.append(i)
    mse_score_RT.append(mean_squared_error(train_labels,tree_prime.predict(train_features.loc[:,flattened])))
    r2_scores_RT.append(r2_score(train_labels, tree_prime.predict(train_features.loc[:,flattened])))
```

### DecisionTreeRegressor MSE, R2 Visualizations

```python
plt.figure(figsize=(16, 9))
plt.xlabel('Number of features selected', fontsize=14, labelpad=20)
plt.ylabel('MSE', fontsize=14, labelpad=20)
plt.plot(features, mse_score_RT)
```

```python
plt.figure(figsize=(16, 9))
plt.xlabel('Number of features selected', fontsize=14, labelpad=20)
plt.ylabel('r2', fontsize=14, labelpad=20)
plt.plot(features, r2_scores_RT)
```

### Comment
Looks like 4-5 features should be enough, lets conduct similar testing with KNearestRegressor

```python
from sklearn.neighbors import KNeighborsRegressor
mse_KNR = []
r2_KNR = []
key_list =[]
for key, value in features_dict.items():
    estimator = KNeighborsRegressor(n_neighbors=128).fit(train_features.loc[:,value], train_labels)
    mse_KNR.append(mean_squared_error(train_labels,estimator.predict(train_features.loc[:,value])))
    r2_KNR.append(r2_score(train_labels, estimator.predict(train_features.loc[:,value])))
    key_list.append(key)
```

### KNeighborsRegressor MSE, R2 Visualizations


```python
plt.figure(figsize=(16, 9))
plt.xlabel('Number of features selected', fontsize=14, labelpad=20)
plt.ylabel('MSE', fontsize=14, labelpad=20)
plt.plot(key_list, mse_KNR)
```

```python
plt.figure(figsize=(16, 9))
plt.xlabel('Number of features selected', fontsize=14, labelpad=20)
plt.ylabel('r2', fontsize=14, labelpad=20)
plt.plot(key_list, r2_KNR)
```

## Selected Features through Feature Engineering

After looking at the plots above, it looks like top 5 features would be the best choice as after that there is very minimal change in r2 and MSE, for both KNR and DTR.

```python
print("Selected Features are \n {}".format(features_dict[5]))
```

```python
train_features = train_features.loc[:, features_dict[5]]
test_features = test_features.loc[:, features_dict[5]]
```

```python
print("Training Features shape is {}".format(train_features.shape))
print("Training Labels shape is {}\n".format(train_labels.shape))
print("Testing Features shape is {}".format(test_features.shape))
print("Testing Labels shape is {}".format(test_labels.shape))
```
